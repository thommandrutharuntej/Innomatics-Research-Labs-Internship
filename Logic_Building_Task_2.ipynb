```python
# Dictionary storing employee names and salaries
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_name = ""
highest_salary = 0
#Loop through each employee in dictionary
for name in employees:
  # Compare current salary with highest_salary
    if employees[name] > highest_salary:
        highest_salary = employees[name]
        highest_name = name

print("Highest Salary:", highest_name, "-", highest_salary)
```

```python
# List of numbers
numbers = [45, 22, 89, 10, 66]

# max and min with first element
max_val = numbers[0]
min_val = numbers[0]

# Loop through the list
for n in numbers:
    # Check for maximum
    if n > max_val:
        max_val = n

    # Check for minimum
    if n < min_val:
        min_val = n

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

```python
# List of product prices
prices = [450, 1200, 899, 1500, 300]

count = 0

# Loop through each price
for p in prices:
    # if price is greater than 1000 or not Check if yes count increases
    if p > 1000:
        count+=1

print("Products above 1000:", count)
```

```python
attendance = ["P", "P", "A", "P", "P"]
 # Count of present days
present = 0
 # Total days
total = 0


for day in attendance:
    total += 1
    if day == "P":
        present += 1

#  percentage calculation
percentage = (present / total) * 100

print("Attendance Percentage:", percentage)
```

```python
s = "python"

freq = {}  # Dictionary for storing the character and its counts

# Loop through each character
for ch in s:
    if ch in freq:
        freq[ch] += 1   # Increases count if character alreasy presents
    else:
        freq[ch] = 1    # Add new character with count of 1

print(freq)
```

```python
marks = [79, 85, 95, 90, 67, 93]
# Calulate the total marks of the student
total = 0
# Average marks of the student
avg = 0
# Loop to calculate the total marks of the student
for mark in marks:
  total+=mark

# Finding Average
avg = total/len(marks)

print(total)
print(avg)
```
