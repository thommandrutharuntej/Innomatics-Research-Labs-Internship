```python
# Function to check whether the recharge amount is valid
def validate_recharge(amount):
    # List of allowed recharge plans
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return False

    if amount not in valid_plans:
        return False

    return True


# Keep asking the user until a valid recharge is entered
while True:
    amt = int(input("Enter recharge amount: "))

    # Validating the amount entered
    if validate_recharge(amt):
        print("Recharge Successful")
        break  # Exit the loop
    else:
        print("Invalid recharge amount. Try again.")
```

```python
# Function to check stock levels of products
def check_stock(products):
    for item, stock in products.items():

        # If stock is below minimum threshold, trigger alert
        if stock < 15:
            print(item, "→ Reorder Alert")
        else:
            print(item, "→ Stock OK")

# (product → available units)
inventory = {
    "Rice": 10,
    "Sugar": 20,
    "Oil": 12
}

check_stock(inventory)
```

```python
# Function to determine student result based on marks
def student_result(marks):
    total = 0
# Calculate the total sum
    for m in marks:
        total += m

    # Calculate average marks
    avg = total / len(marks)

    return "Pass" if avg >= 50 else "Fail"


# Sample marks for a student
marks_list = [45, 60, 70]

# Function call to display the student result
print("Result:", student_result(marks_list))
```

```python
# Function to calculate cab fare
def calculate_fare(km, peak_hour):
    base_fare = 50
    total = base_fare + (12 * km)

    if peak_hour:
        total += total * 0.25

    return total


while True:
    # Get trip details from user
    km = float(input("Enter distance in km: "))
    peak = input("Peak hour? (yes/no): ").lower()

    # Calculate fare
    fare = calculate_fare(km, peak == "yes")
    print("Total Fare: ₹", fare)

    # Ask whether to repeat calculation or not
    retry = input("Calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

```python
# Function to check attendance eligibility
def attendance_check(attendance):
    present = 0  # Count for no.of  present days

    # Count number of days marked as present
    for day in attendance:
        if day == "P":
            present += 1

    # Calculate attendance percentage
    percentage = (present / len(attendance)) * 100



    if percentage>= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# (P = Present, A = Absent)
attendance_list = ["P", "P", "A", "P", "P"]

# Calling the function to display the result
print("Status:", attendance_check(attendance_list))
```

```python
# Function to evaluate password strength
def password_strength(password):
    special_chars = "@#$" # Special Characters

    # Minimum length the password to be
    if len(password) < 8:
        return "Weak Password"

    # Check if password contains at least one digit
    has_digit = any(ch.isdigit() for ch in password)

    # Check if password contains at least one special character
    has_special = any(ch in special_chars for ch in password)

    # Password is strong or not
    return "Strong Password" if has_digit and has_special else "Weak Password"


# Take password input from user
pwd = input("Enter password: ")

# Display the result by calling the function
print(password_strength(pwd))
```
