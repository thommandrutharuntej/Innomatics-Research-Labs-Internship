```python
# Problem 1: Smart Light Controller

# Class that represents a smart light
class SmartLight:

    # This runs when a new SmartLight object is created
    def __init__(self, name):
        self.name = name          # Light name
        self.is_on = False        # Light is OFF by default

    # Function to turn ON light
    def turn_on(self):
        self.is_on = True         # Change state to ON

    # Function to turn the light OFF
    def turn_off(self):
        self.is_on = False        # Change state to OFF

    # Function to display the current light status
    def show_status(self):
        # Check whether the light is ON or OFF
        if self.is_on:
            print(f"{self.name} is ON.")
        else:
            print(f"{self.name} is OFF.")



# Light details from the user
light_name = input("Light Name: ")
action = input("Action (ON / OFF): ")

# Create a SmartLight object by user input
light = SmartLight(light_name)

# Decide what action to perform
if action.upper() == "ON":
    light.turn_on()      # Turn the light ON
elif action.upper() == "OFF":
    light.turn_off()     # Turn the light OFF
else:
    print("Invalid action!")  # Handle wrong input


light.show_status()
```

```python
# Problem 2: Employee ID Card System

# Class that represents an Employee ID Card
class Employee:

    # Constructor to for Employee Details
    def __init__(self, name, emp_id, department):
        self.name = name              # Employee name
        self.emp_id = emp_id          # Employee ID
        self.department = department  # Department name

    # Display method
    def display_id_card(self):
        print("\nEmployee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")



# Input from user
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

# Create an Employee object
employee = Employee(name, emp_id, department)

employee.display_id_card()
```

```python
# Problem 3: Mobile Contact Record

# Class that represents a Contact
class Contact:

    # Constructor for contact details
    def __init__(self, name, phone):
        self.name = name     # Contact name
        self.phone = phone   # Phone number

    # Method to display saved contact information
    def display_contact(self):
        print("\nContact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


# Ccontact details from the user
name = input("Contact Name: ")
phone = input("Phone Number: ")

# Create a Contact object with user input
contact = Contact(name, phone)

contact.display_contact()
```

```python
# Problem 4: Product Price Tag Generator

# Class that represents a Product
class Product:

    # Constructor to store product details
    def __init__(self, name, price):
        self.name = name      # Product name
        self.price = price    # Product price

    # Method to print the price tag
    def print_price_tag(self):
        print(f"\nProduct: {self.name}")
        print(f"Price: ₹{self.price}")


# Product details from user
name = input("Product Name: ")
price = input("Price: ")

# Create a Product object using input
product = Product(name, price)

product.print_price_tag()
```

```python
# Problem 5: Movie Rating Display System

# Class that represents a Movie
class Movie:

    # Constructor to store movie details
    def __init__(self, name, rating):
        self.name = name        # Movie name
        self.rating = rating    # Movie rating

    # Display movie information
    def display_details(self):
        print(f"\nMovie: {self.name}")
        print(f"Rating: {self.rating} / 5")


# Get movie details from user
name = input("Movie Name: ")
rating = input("Rating: ")

# Create Movie object using input
movie = Movie(name, rating)

movie.display_details()
```

```python
# Problem 6: Delivery Address Manager

# Class that represents Delivery Information
class Delivery:

    # Constructor to store customer details
    def __init__(self, customer_name, address):
        self.customer_name = customer_name  # Customer name
        self.address = address              # Delivery address

    # Display method to print delivery details
    def display_details(self):
        print("\nDelivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


# Delivery information from user
name = input("Customer Name: ")
address = input("Address: ")

# Create Delivery object using input data
delivery = Delivery(name, address)

delivery.display_details()
```
