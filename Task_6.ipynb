```python
def smart_parking(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        # Update parked count based on entry/exit
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1

        # Track peak usage
        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    # Capacity check
    if parked > capacity:
        print("Parking Status: Exceeded Capacity")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, logs)
```

```python
def estimate_delivery_time(distance, traffic, weather):
    # Time calculation (assume 5 minutes per km)
    eta = distance * 5

    # Applying traffic delay
    if traffic == "High":
        eta += 15
    elif traffic == "Medium":
        eta += 8
    elif traffic == "Low":
        eta += 3

    # Apply weather delay
    if weather == "Rainy":
        eta += 10
    elif weather == "Stormy":
        eta += 20
    elif weather == "Clear":
        eta += 0

    print("Estimated Delivery Time:", eta, "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```

```python
def analyze_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

    # Suggest additional show if demand is high
    if occupancy >= 90:
        print("Suggestion: Consider opening additional shows")


# Sample Input (150 booked seats)
total_seats = 200
booked_seats = [1] * 150   # Simulating 150 bookings

analyze_occupancy(total_seats, booked_seats)
```

```python
def classify_server_load(readings):
    total = 0

    # Sum all CPU readings
    for r in readings:
        total += r

    avg = total / len(readings)

    print("Average CPU Load:", int(avg), "%")

    # Classify based on rules
    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```python
def monitor_resources(usage):
    overused = []

    for resource, hours in usage.items():
        # Mark resource as overused if usage exceeds threshold
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage_data)
```

```python
def manage_registrations(capacity, total_requests):
    confirmed = 0
    waitlisted = 0

    for i in range(total_requests):
        # Fill confirmed slots until capacity is reached
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    # Registration status check
    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
capacity = 100
registrations = 105

manage_registrations(capacity, registrations)
```
