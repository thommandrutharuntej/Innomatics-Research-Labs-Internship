```python
# Problem 3: Agriculture – Rainfall Adequacy Checker

# Function to check rainfall adequacy
def rainfall_checker(rainfall_data, required_level):
    total = 0

    # Calculate total rainfall
    for rain in rainfall_data:
        total += rain

    # Compute average rainfall
    average = total / len(rainfall_data)

    # Determine rainfall status
    status = "Adequate Rainfall" if average >= required_level else "Inadequate Rainfall"

    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)


# (seasonal rainfall values)
season_rainfall = [60, 80, 70, 78]
required_rainfall = 70

rainfall_checker(season_rainfall, required_rainfall)
```

```python
# Problem 4: Social Media – Duplicate Account Detection

# Function to detect duplicate usernames
def detect_duplicates(usernames):
    has_duplicates = len(usernames) != len(set(usernames))

    print("Duplicate Accounts Found:", "Yes" if has_duplicates else "No")


# Example usernames
accounts = ["user1", "user2", "user3", "user1"]

detect_duplicates(accounts)
```

```python
# Problem 5: Healthcare – Appointment Eligibility Checker

# Function to check appointment eligibility
def check_eligibility(age):
    status = "Eligible" if age >= 18 else "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)


# Example patient age
patient_age = 21

check_eligibility(patient_age)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter

# Function to filter premium crop prices
def filter_premium_prices(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


# Example crop prices
crop_prices = [1500, 2500, 1800, 3200]

filter_premium_prices(crop_prices)
```

```python
# Problem 7: System Monitoring – Application Health Checker

# Function to evaluate application health
def evaluate_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", error_count)
    print("System Status:", status)


# Example input
errors = 7

evaluate_health(errors)
```
