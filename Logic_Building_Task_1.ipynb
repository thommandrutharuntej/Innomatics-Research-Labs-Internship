```python
# Actual username and password
user_name="admin"
password="1234"
# Input for username and password
input_username=input("Enter Username: ")
input_password=input("Enter Password: ")
# Main Logic
if input_username==user_name and input_password==password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
marks=[45,78,90,33,60]
# Counters for pass/fail students
pass_count=0
fail_count=0
# Main Logic
for mark in marks:
    if mark>=50:
        pass_count+=1
    else:
        fail_count+=1

print(f"Total number of students: {len(marks)}")
print(f"Number of students passed: {pass_count}")
print(f"Number of students failed: {fail_count}")
```

```python
names = [" Alice ", "bob", " CHARLIE "]
# Cleaned list: each name stripped and lowercased
cleaned=[]
# Main Logic
for i in names:
    i=i.strip()
    i=i.lower()
    cleaned.append(i)
print(cleaned)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]
size=0;
flag=[]
# Flag messages longer than 10 characters
for i in messages:
    size=len(i)
    print(f"Message: '{i}', Length: {size}")
    if(len(i)>10):
        flag.append(i)
print(f"Messages longer than 10 characters: {flag}")
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0;
# Count errors in logs
for i in logs:
  if(i=="ERROR"):
    count+=1
print(f"Number of errors: {count}")
```
