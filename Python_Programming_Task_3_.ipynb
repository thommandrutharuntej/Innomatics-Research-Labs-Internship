```python
employees={
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# To find the highest score
top_score=max(employees.values())

# Find employees with highest score
top_performers=[]

for name, score in employees.items():
    if score==top_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {top_score})")
```

```python
query = "Buy mobile phone buy phone online"

#Convert to lowercase
query = query.lower()

# Splitting the entire sentence into individual words
words=query.split()

freq={}

#Counting the occurances
for word in words:
    freq[word]=freq.get(word, 0) + 1

repeated_words={}

#Words appeared morethan once
for word, count in freq.items():
    if count>1:
        repeated_words[word] = count

print(repeated_words)
```

```python
sensor=[3, 4, 7, 8, 10, 12, 5]

valid=[]

for hr, val in enumerate(sensor):
    if val%2==0:
        valid.append((hr, val))

print("Valid Sensor Readings (Hour, Value):")
print(valid)
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain={}

#counting occurances
for email in emails:
    dom=email.split("@")[1]
    domain[dom]=domain.get(dom, 0) + 1

total_users = len(emails)

#Calculating the percentage
for domain, count in domain.items():
    perc=(count/total_users)*100
    print(f"{domain}: {perc:.0f}%")
```

```python
sales=[1200, 1500, 900, 2200, 1400, 3000]

# Calculating average sales
avg_sales=sum(sales)/len(sales)

thres=avg_sales*1.3

for day, val in enumerate(sales, start=1):
    if val>thres:
        print(f"Day {day}: {val}")
```

```python
user_ids=["user1", "user2", "user1", "user3", "user1", "user3"]

id_count={}

# Count occurrences of each user ID
for user_id in user_ids:
    id_count[user_id] = id_count.get(user_id, 0) + 1

# Display only duplicate IDs
for user_id, count in id_count.items():
    if count>1:
        print(f"{user_id} → {count} times")
```
